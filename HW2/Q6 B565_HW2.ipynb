```python
import numpy as np
import math
import matplotlib.pyplot as plt
import collections
```

```python
datapoints = []
plot_points = []
```

```python
for i in range(2,51):
    print('Dimension: ', i)
    euclidean_distance = []
    points = []
    for j in range(1000):
        points.append(np.random.uniform(-1,1,i))
    print("total points", len(points))
    for k in points:
        for l in points:
            if collections.Counter(k) == collections.Counter(l):
                continue
            else:
                euclidean_distance.append(np.linalg.norm(k-l))
    minimum = min(euclidean_distance)
    maximum = max(euclidean_distance)
    print('minimum distance = ', minimum)
    print('maximum distance = ', maximum)
    print()
    yplot = math.log10(maximum - minimum)/minimum
    plot_points.append(yplot)
```

```python
plt.plot(range(2,51), plot_points)
plt.ylabel('distance')
plt.xlabel('dimensions')
plt.show()
```
