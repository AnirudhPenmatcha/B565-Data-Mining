# PCA

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
from sklearn.datasets import load_digits
(dataframe, labels) = load_digits(as_frame = True, return_X_y = True)
```

```python
model = PCA(n_components = 2)
model.fit(dataframe)
model_output = model.transform(dataframe)
model_output = pd.DataFrame(model_output,columns = ["dim1", "dim2"], index = range(0, len(dataframe)))
model_output['numbers'] = labels
model_output
```

# t-SNE

```python
model2 = TSNE(n_components = 2)
model2_output = model2.fit_transform(dataframe)
model2_output = pd.DataFrame(model2_output, columns = ['dim1', 'dim2'], index = range(0, len(dataframe)))
model2_output['numbers'] = labels
model2_output
```

```python
colors = {0: 'Red', 1: 'Green', 2: 'Blue', 3: 'Yellow', 4: 'Orange', 5: 'Purple', 6: 'Pink', 7: 'Brown', 8: 'Cyan', 9: 'Magenta'}

markers= {0: 'o', 1: 's', 2: '^', 3: 'd', 4: 'v', 5: '>', 6: '<', 7: 'p', 8: '*', 9: 'H'}
```

```python
for i in markers:
    datapoints = model_output[model_output['numbers']==i]
    plt.scatter(datapoints['dim1'], datapoints['dim2'], c=colors[i], marker=markers[i])
    
```

```python
for i in markers:
    datapoints = model2_output[model2_output['numbers']==i]
    plt.scatter(datapoints['dim1'], datapoints['dim2'], c=colors[i], marker = markers[i])
```
